```python
class_category = ['apple pie', 'baby back ribs', 'baklava', 'beef carpaccio', 'beef tartare', 'beet salad', 'beignets', 'bibimbap', 'bread pudding', 'breakfast burrito', 'bruschetta', 'caesar salad', 'cannoli', 'caprese salad', 'carrot cake', 'ceviche', 'cheese plate', 'cheesecake', 'chicken curry', 'chicken quesadilla', 'chicken wings', 'chocolate cake', 'chocolate mousse', 'churros', 'clam chowder']
```

```python
import requests
import pandas as pd

api_key = 'PUT YOUR API KEY HERE'

def get_recipe_info(recipe):
    url_of_recipe = f'https://api.spoonacular.com/recipes/complexSearch?query={recipe}&number=1&apiKey={api_key}&addRecipeNutrition=true'
    extracting = requests.get(url_of_recipe)

    if extracting.status_code == 200:
        extraction = extracting.json()
        if extraction['results']:
            id_of_recipe = extraction['results'][0]['id']
            url_of_recipe = f'https://api.spoonacular.com/recipes/{id_of_recipe}/information?apiKey={api_key}&includeNutrition=true'
            information_recipe = requests.get(url_of_recipe)

            if information_recipe.status_code == 200:
                return information_recipe.json()
            else:
                print(f"Failed to retrieve recipe info for {recipe}: {information_recipe.status_code}")
                return None
        else:
            print(f"No recipe found for {recipe}")
            return None
    else:
        print(f"Failed to search for {recipe}: {extracting.status_code}")
        return None

nutritional_data = []

for recipe in class_category:
    dish = get_recipe_info(recipe)
    if dish:
        ingredients = ', '.join([ingredient['name'] for ingredient in dish.get('extendedIngredients', [])])
        instructions = dish.get('instructions', 'N/A')  
        info_nutrient = {}
        calories = 0
        if 'nutrition' in dish and 'nutrients' in dish['nutrition']:
            for nutrient in dish['nutrition']['nutrients']:
                info_nutrient[nutrient['name']] = nutrient['amount']
                if nutrient['name'].lower() == 'calories':
                    calories = nutrient['amount']
        else:
            print(f"Nutrient information not available for {recipe}")

        
        nutritional_data.append({
            'Food Category': recipe,  
            'Ingredients': ingredients,
            'Instructions': instructions,
            'Calories': calories,
            **info_nutrient 
        })


df = pd.DataFrame(nutritional_data)
print(df)
df.to_csv('First25_nutrition.csv', index=False)
```
